```python
# Melakukan Coding disini
x = np.array(df)
print (x)
```

**Tulis Codingan ini**

```python
# Melakukan Coding disini
plt.scatter(x[:,0],x[:,1], Label='True position')
plt.xlabel("gaji")
plt.ylabel("pengeluaran")
plt.title("grafik konsumen")
plt.show()
```

**LANGKAH 4**

**MENGGUNAKAN LIBRARY K-MEANS**

**Tulis codingan ini**

image.png

**Tampilan run**

image.png

```python
# Melakukan Coding disini
kmeans = KMeans(n_clusters=7)
kmeans.fit(x)
```

**Tulis Codingan ini**

image.png

**Tampilan run**

image.png

```python
print(kmeans.cluster_centers_)
```

**Tulis Codingan ini**

image.png

**Tampilan run**

image.png

```python
print(kmeans.labels_)
```

**LANGKAH 5**

**MENAMPILKAN OUTPUT**

**Tulis Codingan ini**

image.png

**Tampilan run**

image.png

```python
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_, cmap = 'rainbow')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show
```

**Tulis Codingan ini**

image.png

**Tampilan run**

image.png

```python
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_,
            cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='black')
plt.xlabel("Gji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show()
```
